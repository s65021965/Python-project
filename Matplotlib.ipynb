```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x, y, marker = 'o', linestyle = '-', color = 'b', linewidth = '3')
```

```python
x = [1,2,3,4,5]
y = [6,7,8,9,10]
plt.plot(x, y, marker = 'o', linestyle = '-', color = 'b', linewidth = '3')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title("Line Chart")
plt.grid(True)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y1, label = 'Line 1')
plt.plot(x, y2, label = 'Line 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Two Lines Graph')
plt.legend()
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 2, 3, 4, 5]

plt.plot(x, y1, label = 'Line 1')
plt.plot(x, y2, label = 'Line 2')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')
plt.title('Two Lines Graph')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
data = np.random.randn(1000)   # Histogram is used for continuous numerical data distribution
plt.hist(data, bins = 30, edgecolor = 'black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Pie Chart
Students = ['Amit', 'Mohan', 'Shyam', 'Ramu']
Courses = [5, 3, 4, 1]
plt.pie(Courses, labels = Students, autopct = '%1.1f%%', startangle = 140)
plt.axis('equal')
plt.show()
```
