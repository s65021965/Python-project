```python
print(df['Education'].value_counts())
print(df['City'].value_counts())
print(df['Gender'].value_counts())
print(df['ExperienceInCurrentDomain'].value_counts())
```

```python
df['Gender'].value_counts().plot(kind = 'bar',color = ['blue', 'red'])
plt.title("Employee distribution by Gender")
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
df['Age'].plot(kind = 'hist', bins = 10)
plt.title("Age Distribution of Employees")
plt.xlabel("Age")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
df['City'].value_counts().plot(kind = 'bar')
plt.title("Distribution of Employees across cities")
plt.xlabel("City")
plt.ylabel("Count")
plt.show()
```

```python
pd.crosstab(df['Education'],df['PaymentTier']).plot(kind = 'bar', stacked = True, color = ['yellow','green','red'])
plt.title("Education Level vs PaymentTier")
plt.xlabel('Education')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
subset_df = df[['Age', 'ExperienceInCurrentDomain']]
sns.heatmap(subset_df, annot = True, cmap = 'coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
pd.crosstab(df['EverBenched'], df['LeaveOrNot']).plot(kind='bar', stacked=True, color=['cyan', 'magenta'])
plt.title('Benched Status vs. Leaving the Company')
plt.xlabel('Ever Benched')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x ='EverBenched', y = 'ExperienceInCurrentDomain', data = df)
plt.title('Experience in Current Domain by Benched Status')
plt.show()
```

```python
max_experience = df['ExperienceInCurrentDomain'].max()
plt.figure(figsize=(10, 6))
sns.histplot(df['ExperienceInCurrentDomain'], bins = 20, kde = True)
plt.axvline(max_experience, color='r', linestyle='--', label = f'Max Experience: {max_experience} years')
plt.title('Distribution of Experience Levels in Current Domain')
plt.xlabel('Years of Experience')
plt.ylabel('Number of Employees')
plt.legend()
plt.show()
```
