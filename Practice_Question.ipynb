3. Explain the following terms:
i) The difference in np.array, np.asarray and np.asanyarray
ii) The difference between Deep copy and shallow copy

i) np.array: This function creates a new NumPy array from an iterable (such as a list or tuple). If the input is already a NumPy array, it will make a copy of the input array. This means that it always returns a new copy of the input data, regardless of whether the input is a NumPy array or another array-like object.

np.asarray: This function is similar to np.array, but it has one key difference: if the input is already a NumPy array, it does not create a copy of it. Instead, it returns the input array as is. However, if the input is not a NumPy array, it behaves the same way as np.array and creates a new NumPy array from the input data.

np.asanarray: This function also converts the input to a NumPy array. However, unlike np.array and np.asarray, it does not necessarily create a new array. If the input is already a NumPy array, it returns the input array without making a copy. If the input is an array-like object that can be converted to a NumPy array, it creates a new NumPy array from the input data.

ii) Shallow copy: A shallow copy of an object creates a new object, but does not create copies of nested objects within it. Instead, it copies references to the nested objects. This means that changes made to nested objects in the shallow copy will also affect the original object. In Python, you can create a shallow copy of an object using the copy module's copy() function or the object's copy() method.

Deep copy: A deep copy of an object creates a new object and recursively creates copies of all nested objects within it. This means that changes made to nested objects in the deep copy will not affect the original object, as they are completely independent. In Python, you can create a deep copy of an object using the copy module's deepcopy() function.

```python
#16. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.
np.random.seed(0)
x = np.random.randn(100)
y = np.random.randn(100)

def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'Q1'
    elif x < 0 and y >= 0:
        return 'Q2'
    elif x < 0 and y < 0:
        return 'Q3'
    else:
        return 'Q4'
df = pd.DataFrame({'x': x, 'y': y})
df['Quadrant'] = df.apply(lambda row: get_quadrant(row['x'], row['y']), axis=1)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='deep', style='Quadrant', s=100, markers=['o', 's', 'D', '^'])
plt.axhline(0, color='grey', linestyle='--', linewidth=1)
plt.axvline(0, color='grey', linestyle='--', linewidth=1)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Quadrant-wise Scatter Plot')
plt.legend(title='Quadrant')
plt.grid(True)
plt.show()
```

```python
#17. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
output_notebook()
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)
p = figure(title="Sine Wave Function", x_axis_label='x', y_axis_label='sin(x)', width=800, height=400)
p.line(x, y, legend_label="sin(x)", line_width=2)
show(p)
```

```python
#18. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256

# Generate random data
np.random.seed(0)
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=5)

# Create a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create the bar chart
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="")

# Color bars based on their values
p.vbar(x='categories', top='values', width=0.9, source=source, legend_field="categories",
       line_color='white', fill_color=factor_cmap('categories', palette=Viridis256, factors=categories))

# Add hover tooltips
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"
p.y_range.start = 0
p.xgrid.grid_line_color = None

# Show the plot
output_notebook()
show(p, notebook_handle=True)
```
